# Cdigo de Python mnimo para la visualizacin

Primero se tiene que cargar la libreras necesarias para manejar los datos y graficarlos.

```python
from netCDF4 import Dataset      # Lectura de datos en formato NetCDF
import numpy as np               # Manipulacin de arreglos numricos
import pylab as plt              # Graficacin de datos
```

Cargamos los metadatos del archivo en la variable **archivo**.

Para observar los metadatos se podra introducir su nombre en una casilla de cdigo y observar la salida estndar (Eso no se realiza en este notebook pero usted lo puede realizar para comprenderlo mejor).

```python
archivo = Dataset("/content/OR_SUVI-L1b-Fe284_G16_s20222911802230_e20222911802240_c20222911802435.nc")
```

Ahora se requiere extraer los datos del **archivo**:

En Python el punto que separa palabras indica que la segunda pertenece a la primera. En este caso **variables** est dentro de **archivo**. Finalmente **RAD** es el nombre de la variable que contiene la informacin que estamos buscando. La notacin [:] es de Python y significa lea el contenido completo de principio a fin.

```python
datos = archivo.variables['RAD'][:]
```

Ya que RAD es una matriz de datos podemos obtener una visualizacin rpida y simple de la informacin:

```python
plt.figure(figsize=(10,10),dpi=100)
plt.imshow(datos,cmap='hot',vmin=0,vmax=5)
plt.colorbar()
```
